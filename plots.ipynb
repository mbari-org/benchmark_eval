```python
import matplotlib.pyplot as plt
import numpy as np

cats = np.array(["Simple Midwater", "Simple Benthic", "Difficult Midwater", "Difficult Benthic"])
vals = np.array([82.692, 66.437, 59.491, 68.057])

# bars = plt.bar(cats, vals, color="#4C72B0", edgecolor='black', linewidth=1)
# plt.xlabel("Video Sequence")
# plt.ylabel("HOTA Score")
# plt.tight_layout()
# plt.show()

untuned_vals = [62.557, 50.301, 43.349, 41.216]
x = np.arange(len(cats))  # label locations
width = 0.25  # width of the bars
plt.bar(x - width, untuned_vals, width, label='Before Tuning', color="#e6a222", edgecolor='black', linewidth=0.8)
plt.bar(x, vals, width, label='After Tuning', color='#4C72B0', edgecolor='black', linewidth=0.8)

plt.ylabel('HOTA Score')
plt.xlabel('Video Sequence')
plt.xticks(x, cats)
plt.ylim(0, 100)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
detA = [76.141, 47.554, 45.319, 59.069]
assA = [89.811, 92.855, 78.137, 78.137]
locA = [99.458, 98.407, 97.161, 97.938]

x = np.arange(len(cats))  # label locations
width = 0.25  # width of the bars

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - width, detA, width, label='Detection', color='#168aad', edgecolor='black', linewidth=0.8)
plt.bar(x, assA, width, label='Association', color='#52b69a', edgecolor='black', linewidth=0.8)
plt.bar(x + width, locA, width, label='Localization', color='#d9ed92', edgecolor='black', linewidth=0.8)

# Labels and Titles
plt.ylabel('Score')
plt.xlabel('Video Sequence')
plt.xticks(x, cats)
plt.ylim(0, 100)
plt.legend()

plt.tight_layout()
plt.show()
```
